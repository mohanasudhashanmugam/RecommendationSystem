```python
## 1. Answer

import pandas as pd

df = pd.DataFrame({
    'Age': [25, 32, 47, 51, 62],
    'Salary': [40000, 50000, 60000, 52000, 62000],
    'Purchased': [0, 1, 1, 0, 1]
})

X = df[['Age', 'Salary']]
y = df['Purchased']

# For predicting categories (Classification)
from sklearn.feature_selection import SelectKBest, mutual_info_classif
raw_mi_scores = mutual_info_classif(X, y, random_state=42)
print("raw_mi_scores:",raw_mi_scores)
# Combine with SelectKBest to automatically filter the top K features

k_bouncer = SelectKBest(score_func=mutual_info_classif, k=1)
X_selected = k_bouncer.fit_transform(X, y)

# 4. Identify which column names were kept
selected_indices = k_bouncer.get_support()  ## .get_support() method gives you a True/False mask (a boolean array) that tells  
                                         ##exactly which features were kept and which ones were dropped by your feature selection tool.
chosen_features = X.columns[selected_indices].tolist()

print("\n--- Feature Selection Results ---")
print(f"New X shape: {X_selected.shape}")
print(f"The 1 most useful columns identified: {chosen_features}")
```

```python
##2.Answer
import pandas as pd

data = pd.DataFrame({
    'User': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Item': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Orange'],
    'Rating': [5, 3, 4, 2, 5, 4]
})

sum_ratings = list(data.groupby('Item')['Rating'].sum().head(2).index)
print("Recommend top 2 items to a new user:",sum_ratings)

```

```python
#3.
from sklearn.datasets import load_diabetes
import pandas as pd

X, y = load_diabetes(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using Recursive Feature Elimination (RFE). Use Linear Regression as the estimator and select the top 5 features. Print the selected feature names.

Expected output:
Selected features: ['bmi', 's5', 'bp', 'age', 'sex']
```

```python
#3.Answer

from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression


x, y = load_diabetes(return_X_y=True, as_frame=True)

LR=LinearRegression()

rfe=RFE(estimator=LR,n_features_to_select=5)
rfe_fit=rfe.fit(x,y)
#rfe_feature=rfe_fit.transform(x)

# Extract the selected feature names using the support_ = which is same as .get_support to get boolean mask - True/false
selected_features = x.columns[rfe.support_].tolist()  

print("Selected features:",selected_features)

```

```python
#4.Answer

from sklearn.datasets import load_digits
import pandas as pd

X, y = load_digits(return_X_y=True, as_frame=True)


from sklearn.feature_selection import SelectKBest, mutual_info_classif

k_bouncer = SelectKBest(score_func=mutual_info_classif, k=8)
X_selected = k_bouncer.fit_transform(X, y)

chosen_features = X.columns[k_bouncer.get_support()].tolist()

print(f"Top 8 features using mutual info: {chosen_features}")

```

```python
#5.
from sklearn.datasets import load_breast_cancer
import pandas as pd

X, y = load_breast_cancer(return_X_y=True, as_frame=True)
Write a Python program to perform feature selection using SelectFromModel. Train a Logistic Regression model and select features based on importance using SelectFromModel. Print the names of the selected feature


Expected Output:
Selected features: ['mean radius', 'mean concavity', 'texture error', 'worst texture', 'worst smoothness', 'worst compactness', 'worst concavity', 'worst concave points', 'worst symmetry']
```

```python
#5.Answer
from sklearn.datasets import load_breast_cancer
import pandas as pd
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression

x, y = load_breast_cancer(return_X_y=True, as_frame=True)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/4,random_state=0)


# 1. Initialize the base model
LR = LogisticRegression(penalty='l1', solver='liblinear', random_state=42)

# 2. Wrap it with SelectFromModel
selector = SelectFromModel(estimator=LR, threshold="mean")

# 3. Fit to data and transform
selector.fit(x_train, y_train)

selected_features = selector.get_feature_names_out(input_features=x_train.columns)
print("Selected features:" ,list(selected_features) )
```
